```python
def get_shop_info_summary(categoriesID, cityId, page, limit=24):
    url = f"https://mocha.lozi.vn/v6.1/search/eateries?categories={categoriesID}&limit={limit}&page={page}&superCategoryId=1&cityId={cityId}"

    payload = {}
    headers = {
    'accept': '*/*',
    'accept-language': 'vi_VN',
    'origin': 'https://loship.vn',
    'priority': 'u=1, i',
    'referer': 'https://loship.vn/',
    'sec-ch-ua': '"Not/A)Brand";v="8", "Chromium";v="126", "Google Chrome";v="126"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"',
    'sec-fetch-dest': 'empty',
    'sec-fetch-mode': 'cors',
    'sec-fetch-site': 'cross-site',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36',
    'x-access-token': 'unknown',
    'x-city-id': '1',
    'x-lozi-client': '1'
    }

    response = requests.request("GET", url, headers=headers, data=payload)

    data_json = json.loads(response.text)
    return data_json['data'], data_json['pagination']
```

```python
def get_menu_info(slug):
    url = f"https://mocha.lozi.vn/v6.1/eateries/slug:{slug}/menu"

    payload = {}
    headers = {
    'accept': '*/*',
    'accept-language': 'vi_VN',
    'origin': 'https://loship.vn',
    'priority': 'u=1, i',
    'referer': 'https://loship.vn/',
    'sec-ch-ua': '"Not/A)Brand";v="8", "Chromium";v="126", "Google Chrome";v="126"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"',
    'sec-fetch-dest': 'empty',
    'sec-fetch-mode': 'cors',
    'sec-fetch-site': 'cross-site',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36',
    'x-access-token': 'unknown',
    'x-city-id': '50',
    'x-language': 'vi',
    'x-lozi-client': '1'
    }

    response = requests.request("GET", url, headers=headers, data=payload)

    data_json = json.loads(response.text)
    return data_json['data']
```

```python
# pipeline 
citites_configs = func_get_congigs_cities() # l·∫•y th√¥ng tin c√°c th√†nh ph·ªë

# V√≤ng l·∫∑p c√°c th√†nh ph·ªë
for city in citites_configs:
    print("City ID : ", city['city']['id'])
    print("City Name : ", city['city']['name'])
    print("City slug: ", city['city']['slug'])
    cates = func_get_cates_from_cityid(cityId=city['city']['id']) # l·∫•y th√¥ng tin c√°c ng√†nh h√†ng

    # v√≤ng l·∫∑p c√°c ng√†nh h√†ng trong 1 th√†nh ph·ªë
    for cate in cates:
        print(cate['id'])

        # T√≠nh xem xem l√† 1 ng√†nh h√†ng c√≥ bao nhi√™u trang (pages)
        _, data_pages = get_shop_info_summary(categoriesID=cate['id'], cityId=city['city']['id'], page=1, limit=24)
        pages = data_pages['total']//data_pages['limit'] + 1

        # v√≤ng l·∫∑p to√†n c√°c trang trong 1 ng√†nh h√†ng
        for num_page in range(0, pages):
            data_shops, _ = get_shop_info_summary(categoriesID=cate['id'], cityId=city['city']['id'], 
                                                  page=num_page, limit=24)
            
            document_shops = []
            # v√≤ng l·∫∑p t·∫•t c·∫£ c√°c qu√°n ƒÉn trong 1 trang (page)
            for i, shop in enumerate(data_shops):
                print(i ,"slug : ", shop['slug'])
                shop_info = get_shop_full_info(slug=shop['slug']) # l·∫•y to√†n b·ªô th√¥ng tin c·ªßa 1 qu√°n ƒÉn theo th·∫ª slug
                menu_info = get_menu_info(slug=shop['slug']) # l·∫•y to√†n th√¥ng tin c·ªßa menu 
                print("shop_info : ", shop_info)
                print("menu info : ", menu_info)
                print()

                # Th√™m v√†o danh s√°ch c√°c qu√°n ƒÉn
                doc_shop = {
                    'city': city['city'],
                    'category': cate,  
                    'shop_info': shop_info,
                    'menu_info': menu_info, 
                    'createAt': time.time()
                }

                document_shops.append(doc_shop)
            
            # ki·ªÉm tra n·∫øu c√≥ d·ªØ li·ªáu v·ªÅ qu√°n ƒÉn th√¨ m·ªõi insert
            if len(document_shops) > 0:
                # l∆∞u d·ªØ li·ªáu qu√°n ƒÉn
                x = mycol.insert_many(document_shops)
                print(x.inserted_ids)
                print(">>>>> L∆∞u tr·ªØ d·ªØ li·ªáu qu√°n ƒÉn th√†nh c√¥ng <<<<<")
            break
        break
    break
```
