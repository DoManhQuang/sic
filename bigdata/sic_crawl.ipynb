```python
# pipeline 
citites_configs = func_get_congigs_cities() # lấy thông tin các thành phố

# Vòng lặp các thành phố
for idx_city, city in enumerate(citites_configs):
    if city['city']['id'] != 1:
        continue
    print("City ID : ", city['city']['id'])
    print("City Name : ", city['city']['name'])
    print("City slug: ", city['city']['slug'])
    cates = func_get_cates_from_cityid(cityId=city['city']['id']) # lấy thông tin các ngành hàng

    # vòng lặp các ngành hàng trong 1 thành phố
    for cate in cates:
        print(cate['id'])

        # Tính xem xem là 1 ngành hàng có bao nhiêu trang (pages)
        _, data_pages = get_shop_info_summary(categoriesID=cate['id'], cityId=city['city']['id'], page=1, limit=24)
        pages = data_pages['total']//data_pages['limit'] + 1

        # vòng lặp toàn các trang trong 1 ngành hàng
        for num_page in range(0, pages):
            data_shops, _ = get_shop_info_summary(categoriesID=cate['id'], cityId=city['city']['id'], 
                                                  page=num_page, limit=24)
            
            document_shops = []
            # vòng lặp tất cả các quán ăn trong 1 trang (page)
            for i, shop in enumerate(data_shops):
                print(i ,"slug : ", shop['slug'])
                shop_info = get_shop_full_info(slug=shop['slug']) # lấy toàn bộ thông tin của 1 quán ăn theo thẻ slug
                menu_info = get_menu_info(slug=shop['slug']) # lấy toàn thông tin của menu 
                print("shop_info : ", shop_info)
                print("menu info : ", menu_info)
                print()

                # Thêm vào danh sách các quán ăn
                doc_shop = {
                    'city': city['city'],
                    'category': cate,  
                    'shop_info': shop_info,
                    'menu_info': menu_info, 
                    'createAt': time.time()
                }

                document_shops.append(doc_shop)
            
            # kiểm tra nếu có dữ liệu về quán ăn thì mới insert
            if len(document_shops) > 0:
                # lưu dữ liệu quán ăn
                x = mycol.insert_many(document_shops)
                print(x.inserted_ids)
                print(">>>>> Lưu trữ dữ liệu quán ăn thành công <<<<<")

                # đẩy dữ liệu lên kafka
                producer.send("sic-kafka", bytes(str(document_shops), 'utf-8'))
                print(">>>> Đẩy dữ liệu lên Kafka Thành công <<<<")
            break
        break
    break

```

```python
consumer = KafkaConsumer(bootstrap_servers="localhost:9092")
consumer.subscribe('sic-kafka')
```

```python
# last message
partition = TopicPartition('sic-kafka', 0)
end_offset = consumer.end_offsets([partition])
consumer.seek(partition,list(end_offset.values())[0]-1)

for m in consumer:
  print(m.value)
```
